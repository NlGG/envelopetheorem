```python
#Y=(LK)**(1/2) w=5 r=20 Y=20x
import matplotlib.pyplot as plt
import numpy as np
fig, ax = plt.subplots()
x = np.linspace(0, 60, 200)
plt.ylim(0,1500)
plt.title("STC's envelope STC")
plt.xlabel("yield")
plt.ylabel("cost")
for i in range(1,100):
    y = 100+(5*(x**2/i))+i*20
    ax.plot(x, y, 'b-', linewidth=2)
plt.show()
```
