```python
#Y=(LK)**(1/2) (w=5 r=20 100 ) Y=20x+100 
import matplotlib.pyplot as plt
import numpy as np
fig, ax = plt.subplots()
x = np.linspace(0, 60, 200)
plt.ylim(0,1500)
plt.title("STC's envelope")
plt.xlabel("yield")
plt.ylabel("cost")
y=20*x+100
ax.plot(x,y,'r-',linewidth=5)
for i in range(1,100):
    y = 100+(5*(x**2/i))+i*20
    ax.plot(x, y, 'b-', linewidth=2)
plt.show()
```

```python
# Y=(LK)**(1/4)( w=5 r=20 100 )Y=20x**2+100
import matplotlib.pyplot as plt
import numpy as np
fig, ax = plt.subplots()
x = np.linspace(0, 10, 200)
plt.ylim(0,1500)
plt.title("STC's envelope")
plt.xlabel("yield")
plt.ylabel("cost")
y=100+20*x**2
ax.plot(x, y, 'r-', linewidth=5)
for i in range(1,100):
    y = 100+(5*(x**4/i))+i*20
    ax.plot(x, y, 'b-', linewidth=2)
plt.show()
```

```python
# Y=LK (w=5 r=20 100) Y=20x**(1/2)+100
import matplotlib.pyplot as plt
import numpy as np
fig, ax = plt.subplots()
x = np.linspace(0, 1000, 200)
plt.ylim(0,1500)
plt.title("STC's envelope")
plt.xlabel("yield")
plt.ylabel("cost")
y=100+20*(x**(0.5))
ax.plot(x, y, 'r-', linewidth=5)
for i in range(1,100):
    y = 100+(5*(x/i))+i*20
    ax.plot(x, y, 'b-', linewidth=2)
plt.show()
```
